```python
import random
import json

# generating delegation testdata

def tevk_name_gen():
    """returns a random fictive settlement name string"""
    tevk_bits = {
        "start":[
            "Abrak","Béla","Czibak","Dörgő","Enyém","Felső","Gütyök","Random",
            "Hibbant","Izaura","János","Kappan","Lebbencs","Matyó","Nemez",
            "Oppá","Prücsök","Rózsa","Sajdul","Szeben","Túlzó","Tyalak"],
        "mid":[
            "huk","lac","pat","zsib","tornác","fel","alsó","felső","új","",""],
        "end":[
            "háza","háza","falva","falva","ony","tornya","szalók","tány",
            "nyő","óca","oda","út","lak","lak","kereszt","kereszt","lőd",
            "bánya", "fa", "kovácsi","vár","fő", "",""]}
    return (
        random.choice(tevk_bits["start"])+
        random.choice(tevk_bits["mid"])+
        random.choice(tevk_bits["end"])
    )


def delegees_gen():
    """returns a static delegee array"""
    return [
              {
                "userID": "1432533",
                "name": "Ifj. Prof. Nagyonhosszúnevű-Tesztadat Zebulon",
                "szsz": "1 234567 8912",
                "address": "5539 Körösnagyharsány, Kossuth Lajos u. 3",
                "email": "example@gmail.com",
                "phone": "30-873-1234"
              },
              {
                "userID": "1432534",
                "name": "Árvíztűrő Tükörfúrógép",
                "address": "4172 Biharnagybajom, Bajcsy-Zsilinszky utca 5-1.",
                "szsz": "1 234567 8912",
                "email": "example@gmail.com",
                "phone": "30-873-1234"
              }
            ]

def zones_gen():
    """generates 1-5 zones with delegees"""
    return [
        {
            "id":f"{zone+1:02}",
            "delegees":delegees_gen()
        } for zone in range(0,random.randint(1,5))
    ]

def tevks_gen(count=6):
    """generating 6 settlements per OEVK just to get cca 500 docs eventually"""
    for tevk_ix in range(0, count):
        yield {
            "id":f"{tevk_ix+1:03}",
            "name":tevk_name_gen(),
            "zones":zones_gen()}
    

def oevk_gen():
    oevks = [
        #("01", "Budapest", 18), -- this logic does not work here
        ("02", "Baranya", 4),
        ("03", "Bács-Kiskun", 6),
        ("04", "Békés", 4),
        ("05", "Borsod-Abaúj-Zemplén", 7),
        ("06", "Csongrád", 4),
        ("07", "Fejér", 5),
        ("08", "Győr-Moson-Sopron", 5),
        ("09", "Hajdú-Bihar", 6),
        ("10", "Heves", 3),
        ("11", "Jász-Nagykun-Szolnok", 4),
        ("12", "Komárom-Esztergom", 3),
        ("13", "Nógrád", 2),
        ("14", "Pest", 12),
        ("15", "Somogy", 4),
        ("16", "Szabolcs-Szatmár-Bereg", 6),
        ("17", "Tolna", 3),
        ("18", "Vas", 3),
        ("19", "Veszprém", 4),
        ("20", "Zala", 3)
    ]
    
    for county in oevks:
        for oevk in range(0, county[2]):
            yield {
                "id":county[0]+f"-{oevk+1:03}", # nn-nnn format
                "name":county[1]+f" {oevk+1:02}",
                "tevks":list(tevks_gen())
            }


def testdata_gen(filename='../input/user_testdata.json'):
    with open(filename, 'w', encoding='utf-8') as f:
                f.write(
                    str(
                        list(
                            oevk_gen()
                        )
                    ).replace("'",'"')
                )

testdata_gen()
```
