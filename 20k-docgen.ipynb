```python
import json
from jinja2 import Template
from IPython.core.display import display, HTML

def read_users(file_str='../input/test-json-user-data/users.json'):
    """reads the user json dump file and returns as json variable"""
    with open(file_str, 'r', encoding='utf-8') as f:
        oevks = json.load(f)
    return oevks

def read_letter_template(file_str='../input/test-html-template/megbizolevel_teszt.html'):
    """reads the mail template file and returns it as text variable"""
    with open(file_str, 'r', encoding='utf-8') as f:
        template = Template(f.read())
    return template

db = read_users()
t = read_letter_template()

for oevk in read_users():
    for tevk in oevk['tevks']:
        filename = f"\n\n szszb_megbizo_OEVK{oevk['id']}_{oevk['name']}_TEVK{tevk['id']}_{tevk['name']}.pdf \n\n"
        print(filename)
        result = t.render(tevk = tevk)
        display(HTML(result))
```
